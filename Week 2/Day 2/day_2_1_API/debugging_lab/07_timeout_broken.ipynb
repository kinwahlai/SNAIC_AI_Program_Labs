# Challenge 07: Request May Timeout

**PPT problem:** Request may timeout  
**Technique:** Timeout handling / retry

Timeouts prevent a client from waiting forever. But a timeout is also a failure mode that should be handled. A robust client catches requests.Timeout, waits, and retries when appropriate.

````python
"""
Challenge 07: Network timeout.

The server is slow. This client uses a very short timeout and has no retry
strategy, so the job stops immediately.
"""

import pandas as pd
import requests


BASE_URL = "http://127.0.0.1:8011"
CLIENT_ID = "group_07_timeout"


def main() -> None:
    response = requests.get(
        f"{BASE_URL}/debug-lab/07-timeout/taxi/records",
        params={
            "page": 1,
            "page_size": 10,
            "client_id": CLIENT_ID,
        },
        timeout=2,
    )
    response.raise_for_status()

    dataframe = pd.DataFrame(response.json()["data"])
    print("Rows collected:", len(dataframe))
    print(dataframe.head())


if __name__ == "__main__":
    main()
````
