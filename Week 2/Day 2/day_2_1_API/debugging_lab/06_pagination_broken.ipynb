# Challenge 06: Response Data May Be Too Large

**PPT problem:** Response data may be too large  
**Technique:** Pagination

Large APIs split results into pages. A client that only reads page 1 silently misses most data. A robust collector loops until next_page is None or until a cursor is exhausted.

````python
"""
Challenge 06: Pagination.

The API returns only one page at a time. This script wrongly assumes the first
response contains all taxi records.
"""

import pandas as pd
import requests


BASE_URL = "http://127.0.0.1:8011"
CLIENT_ID = "group_06_pagination"


def main() -> None:
    response = requests.get(
        f"{BASE_URL}/debug-lab/06-pagination/taxi/records",
        params={
            "page": 1,
            "page_size": 20,
            "client_id": CLIENT_ID,
        },
        timeout=10,
    )
    response.raise_for_status()
    payload = response.json()

    dataframe = pd.DataFrame(payload["data"])
    print("Rows collected:", len(dataframe))
    print("Total records reported by API:", payload["total_records"])
    print("Expected: rows collected should eventually match total_records.")
    print(dataframe[["record_id", "longitude", "latitude"]].head())


if __name__ == "__main__":
    main()
````
