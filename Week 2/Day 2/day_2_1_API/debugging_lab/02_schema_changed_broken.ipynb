````python
"""
Challenge 02: Schema changed.

The provider renamed one field. The request succeeds, but the processing code
breaks because it still expects the old schema.
"""

import pandas as pd
import requests


BASE_URL = "http://127.0.0.1:8011"
CLIENT_ID = "group_02_schema"


def main() -> None:
    response = requests.get(
        f"{BASE_URL}/debug-lab/02-schema-change/taxi/records",
        params={
            "page": 1,
            "page_size": 20,
            "client_id": CLIENT_ID,
        },
        timeout=10,
    )
    response.raise_for_status()

    payload = response.json()
    dataframe = pd.DataFrame(payload["data"])

    total_taxis = int(dataframe["available_taxi_count"].sum())
    print("Rows:", len(dataframe))
    print("Total available taxis in this page:", total_taxis)
    print(dataframe[["record_id", "available_taxi_count"]].head())


if __name__ == "__main__":
    main()
````
