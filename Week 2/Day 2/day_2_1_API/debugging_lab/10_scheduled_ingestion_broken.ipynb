````python
"""
Challenge 10: Scheduled ingestion.

This script collects only one snapshot, then tries to reason about a trend.
One snapshot is not enough. Learners should collect multiple snapshots over
time and store the collection_time for each one.
"""

import pandas as pd
import requests


BASE_URL = "http://127.0.0.1:8011"
CLIENT_ID = "group_10_scheduled"


def main() -> None:
    response = requests.get(
        f"{BASE_URL}/debug-lab/10-scheduled/taxi/current",
        params={"client_id": CLIENT_ID},
        timeout=10,
    )
    response.raise_for_status()
    payload = response.json()

    dataframe = pd.DataFrame(
        [
            {
                "collection_time": payload["collection_time"],
                "snapshot_rows": payload["snapshot_rows"],
            }
        ]
    )

    print(dataframe)
    assert len(dataframe) >= 3, "Need at least 3 snapshots to discuss a trend."


if __name__ == "__main__":
    main()
````
