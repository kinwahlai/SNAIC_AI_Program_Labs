# Challenge 01: API Requires Permission

**PPT problem:** API requires permission  
**Technique:** Authentication headers / query parameters

APIs often require a key or token. The key can be sent in different places: an HTTP header, a query parameter, a cookie, or a signed request. If the provider changes the authentication contract, old clients may receive 401 Unauthorized even though the key value is still correct.

````python
"""
Challenge 01: Authentication moved.

This script used to work when the API key was accepted in the X-API-Key header.
The provider changed the contract. Run the script, read the error, then fix the
request.
"""

import pandas as pd
import requests


BASE_URL = "http://127.0.0.1:8011"
CLIENT_ID = "group_01_auth"
API_KEY = "class-demo-key"


def main() -> None:
    response = requests.get(
        f"{BASE_URL}/debug-lab/01-auth-moved/taxi/records",
        params={
            "page": 1,
            "page_size": 10,
            "client_id": CLIENT_ID,
        },
        headers={
            "X-API-Key": API_KEY,
        },
        timeout=10,
    )
    response.raise_for_status()

    payload = response.json()
    dataframe = pd.DataFrame(payload["data"])
    print("Taxi records collected:", len(dataframe))
    print(dataframe[["record_id", "longitude", "latitude"]].head())


if __name__ == "__main__":
    main()
````
