## 2. JSON Object

A JSON object is written using curly braces `{}`.

It contains key-value pairs.

Example:

```json
{
  "region": "Central",
  "taxi_count": 120,
  "weather": "Cloudy"
}
```

In Python, a JSON object is usually converted into a dictionary.

````python
import json
import pandas as pd
from pprint import pprint
````

````python
json_text = '''
{
  "region": "Central",
  "taxi_count": 120,
  "weather": "Cloudy"
}
'''

data = json.loads(json_text)

print(type(data))
pprint(data)
````

## 3. JSON Array

A JSON array is written using square brackets `[]`.

Example:

```json
[
  "Central",
  "East",
  "West"
]
```

In Python, a JSON array is usually converted into a list.

````python
json_text = '''
[
  "Central",
  "East",
  "West"
]
'''

data = json.loads(json_text)

print(type(data))
pprint(data)
````

## 4. Object vs Array

This is one of the most important beginner concepts.

| JSON Structure | Syntax | Python Type | Meaning |
|---|---|---|---|
| Object | `{}` | `dict` | One structured record |
| Array | `[]` | `list` | A collection of values or records |

Example:

```json
{
  "station": "S001",
  "temperature": 30.5
}
```

This is one object.

```json
[
  {"station": "S001", "temperature": 30.5},
  {"station": "S002", "temperature": 31.2}
]
```

This is an array of objects.

````python
json_text = '''
[
  {"station": "S001", "temperature": 30.5},
  {"station": "S002", "temperature": 31.2},
  {"station": "S003", "temperature": 29.8}
]
'''

weather_data = json.loads(json_text)

print(type(weather_data))
print(type(weather_data[0]))
pprint(weather_data[0])
````

## 5. JSON Data Types

JSON supports a small set of data types.

| JSON Type | Example | Python Type |
|---|---|---|
| String | `"Singapore"` | `str` |
| Number | `123`, `30.5` | `int`, `float` |
| Boolean | `true`, `false` | `True`, `False` |
| Null | `null` | `None` |
| Object | `{ "a": 1 }` | `dict` |
| Array | `[1, 2, 3]` | `list` |

Important differences between JSON and Python:

| JSON | Python |
|---|---|
| `true` | `True` |
| `false` | `False` |
| `null` | `None` |
| double quotes required | Python allows single or double quotes for strings |

````python
json_text = '''
{
  "city": "Singapore",
  "temperature": 30.5,
  "rain": false,
  "warning": null,
  "forecast": ["Cloudy", "Rain", "Fair"],
  "metadata": {
    "source": "sample_api",
    "unit": "Celsius"
  }
}
'''

data = json.loads(json_text)

for key, value in data.items():
    print(key, "=>", value, "| Python type:", type(value))
````

## 6. Valid JSON vs Invalid JSON

JSON has strict syntax rules.

Common rules:

1. Keys must be strings with double quotes.
2. String values must use double quotes.
3. No trailing comma.
4. Comments are not allowed.
5. Boolean values are lowercase: `true`, `false`.
6. Null value is lowercase: `null`.

The following is valid JSON:

```json
{
  "name": "Alice",
  "age": 25
}
```

The following is invalid JSON:

```json
{
  name: "Alice",
  "age": 25,
}
```

Problems:

- `name` has no double quotes.
- There is a trailing comma after `25`.

````python
# Valid JSON

valid_json = '''
{
  "name": "Alice",
  "age": 25
}
'''

data = json.loads(valid_json)
pprint(data)
````

````python
# Invalid JSON example
# Run this cell and observe the error message.

invalid_json = '''
{
  name: "Alice",
  "age": 25,
}
'''

try:
    data = json.loads(invalid_json)
except json.JSONDecodeError as e:
    print("JSONDecodeError:", e)
````

## 7. Python `json` Library

Python has a built-in library called `json`.

Common functions:

| Function | Purpose |
|---|---|
| `json.loads()` | Convert JSON string to Python object |
| `json.dumps()` | Convert Python object to JSON string |
| `json.load()` | Read JSON from a file |
| `json.dump()` | Write JSON to a file |

Remember:

| Function | Input | Output |
|---|---|---|
| `loads` | JSON string | Python object |
| `dumps` | Python object | JSON string |
| `load` | File object | Python object |
| `dump` | Python object + file object | JSON file |

## 8. `json.loads()` and `json.dumps()`

`loads` means "load from string".

`dumps` means "dump to string".

````python
json_text = '''
{
  "course": "Data Engineering",
  "day": 1,
  "topics": ["API", "JSON", "Parquet"]
}
'''

python_obj = json.loads(json_text)

print("After json.loads():")
print(type(python_obj))
pprint(python_obj)

json_string = json.dumps(python_obj, indent=2)

print("\nAfter json.dumps():")
print(type(json_string))
print(json_string)
````

### Note

`indent=2` makes the JSON output easier to read.

Without indentation, JSON is still valid but less readable.

````python
compact_json = json.dumps(python_obj)
pretty_json = json.dumps(python_obj, indent=2)

print("Compact JSON:")
print(compact_json)

print("\nPretty JSON:")
print(pretty_json)
````

## 9. `json.load()` and `json.dump()`

`load` and `dump` are used with files.

Let us save a Python object to a JSON file and read it back.

````python
course_config = {
    "course_name": "Data Engineering",
    "duration_days": 5,
    "topics": [
        "API ingestion",
        "JSON",
        "Parquet",
        "SQL",
        "Airflow"
    ],
    "has_project": True
}

with open("course_config.json", "w", encoding="utf-8") as f:
    json.dump(course_config, f, indent=2)

print("Saved course_config.json")
````

````python
with open("course_config.json", "r", encoding="utf-8") as f:
    loaded_config = json.load(f)

pprint(loaded_config)
````

## 11. Nested JSON

Real API responses are often nested.

Example:

```json
{
  "timestamp": "2026-05-26T09:00:00",
  "region": "Central",
  "weather": {
    "condition": "Cloudy",
    "temperature": 30.5,
    "humidity": 78
  }
}
```

The `weather` field is another object inside the main object.

````python
json_text = '''
{
  "timestamp": "2026-05-26T09:00:00",
  "region": "Central",
  "weather": {
    "condition": "Cloudy",
    "temperature": 30.5,
    "humidity": 78
  }
}
'''

data = json.loads(json_text)

print("Condition:", data["weather"]["condition"])
print("Temperature:", data["weather"]["temperature"])
````

## 13. Nested Array Example: Taxi Locations

Many real APIs return nested arrays.

For example, a taxi availability API may return:

```json
{
  "timestamp": "2026-05-26T09:00:00",
  "taxis": [
    {"lat": 1.3001, "lon": 103.8001},
    {"lat": 1.3102, "lon": 103.8102}
  ]
}
```

This is not a flat table yet.

````python
taxi_json = '''
{
  "timestamp": "2026-05-26T09:00:00",
  "source": "sample_taxi_api",
  "taxis": [
    {"lat": 1.3001, "lon": 103.8001},
    {"lat": 1.3102, "lon": 103.8102},
    {"lat": 1.3203, "lon": 103.8203}
  ]
}
'''

taxi_data = json.loads(taxi_json)

rows = []

for taxi in taxi_data["taxis"]:
    rows.append({
        "timestamp": taxi_data["timestamp"],
        "source": taxi_data["source"],
        "latitude": taxi["lat"],
        "longitude": taxi["lon"]
    })

df_taxi = pd.DataFrame(rows)
df_taxi
````

## 14. Flattening JSON with `pandas.json_normalize()`


`pd.json_normalize()` is a useful pandas function for converting nested JSON into a flatter table format.

### Key Idea

In many cases， if the input is a list of JSON objects, then usually:

> each JSON object becomes one row.

Nested fields can be expanded into separate columns.

For example, a nested field such as:

`weather.condition`

can become a column in the table.

### Note

`json_normalize()` helps flatten nested JSON, but it does not decide the meaning of the data for us.

Before converting JSON into a table, we still need to ask:

> What should one row represent?



````python
weather_records = [
    {
        "region": "Central",
        "weather": {
            "condition": "Cloudy",
            "temperature": 30.5,
            "humidity": 78
        }
    },
    {
        "region": "East",
        "weather": {
            "condition": "Rain",
            "temperature": 28.9,
            "humidity": 85
        }
    }
]

df_weather = pd.json_normalize(weather_records)
df_weather
````

Notice how nested fields become columns such as:

```text
weather.condition
weather.temperature
weather.humidity
```

## 15. JSON Lines / NDJSON

Another important format is **JSON Lines**, also called **NDJSON**.

Instead of one big JSON array, each line is a separate JSON object.

Example:

```json
{"user_id": 1, "event": "click"}
{"user_id": 2, "event": "purchase"}
{"user_id": 3, "event": "logout"}
```

This is common in logs and streaming data.

Why it is useful:

| Reason | Explanation |
|---|---|
| Streaming-friendly | Can process one line at a time |
| Large-file friendly | No need to load the whole file |
| Log-friendly | Each event is one line |
| Easy to append | Add new events at the end |

````python
json_lines_text = '''
{"user_id": 1, "event": "click"}
{"user_id": 2, "event": "purchase"}
{"user_id": 3, "event": "logout"}
'''.strip()

records = []

for line in json_lines_text.splitlines():
    records.append(json.loads(line))

df_events = pd.DataFrame(records)
df_events
````

## 16. GeoJSON

GeoJSON is a special JSON-based format for geospatial data.

It is commonly used to represent:

- Points
- Lines
- Polygons
- MultiPolygons
- Geospatial features
- Feature collections

GeoJSON is useful in data engineering when we work with maps, coordinates, regions, and spatial joins.

Example: a point location in GeoJSON:

```json
{
  "type": "Point",
  "coordinates": [103.8198, 1.3521]
}
```

Important:

> GeoJSON coordinates are usually written as `[longitude, latitude]`, not `[latitude, longitude]`.

Example: a simple region boundary in GeoJSON:

```json
{
  "type": "Feature",
  "properties": {
    "name": "Tampines"
  },
  "geometry": {
    "type": "Polygon",
    "coordinates": [
      [
        [103.9200, 1.3300],
        [103.9700, 1.3300],
        [103.9700, 1.3700],
        [103.9200, 1.3700],
        [103.9200, 1.3300]
      ]
    ]
  }
}

## 17. GeoJSON Feature

A GeoJSON feature combines geometry with properties.

Example:

```json
{
  "type": "Feature",
  "geometry": {
    "type": "Point",
    "coordinates": [103.8198, 1.3521]
  },
  "properties": {
    "name": "Singapore",
    "category": "city"
  }
}
```

````python
geojson_point = {
    "type": "Point",
    "coordinates": [103.8198, 1.3521]
}

print("GeoJSON type:", geojson_point["type"])
print("Longitude:", geojson_point["coordinates"][0])
print("Latitude:", geojson_point["coordinates"][1])
````

````python
geojson_feature = {
    "type": "Feature",
    "geometry": {
        "type": "Point",
        "coordinates": [103.8198, 1.3521]
    },
    "properties": {
        "name": "Singapore",
        "category": "city"
    }
}

print("Feature type:", geojson_feature["type"])
print("Geometry type:", geojson_feature["geometry"]["type"])
print("Name:", geojson_feature["properties"]["name"])
print("Longitude:", geojson_feature["geometry"]["coordinates"][0])
print("Latitude:", geojson_feature["geometry"]["coordinates"][1])
````

## 18. GeoJSON FeatureCollection

A FeatureCollection contains multiple features.

This is very common when representing many locations or many regions.

Example use cases:

| Use Case | GeoJSON Geometry |
|---|---|
| Taxi locations | Point |
| Road segments | LineString |
| Planning areas | Polygon |
| Country boundaries | MultiPolygon |

````python
geojson_feature_collection = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "geometry": {
                "type": "Point",
                "coordinates": [103.8001, 1.3001]
            },
            "properties": {
                "taxi_id": "T001",
                "region": "Central"
            }
        },
        {
            "type": "Feature",
            "geometry": {
                "type": "Point",
                "coordinates": [103.8102, 1.3102]
            },
            "properties": {
                "taxi_id": "T002",
                "region": "East"
            }
        }
    ]
}

rows = []

for feature in geojson_feature_collection["features"]:
    coordinates = feature["geometry"]["coordinates"]
    properties = feature["properties"]

    rows.append({
        "taxi_id": properties["taxi_id"],
        "region": properties["region"],
        "longitude": coordinates[0],
        "latitude": coordinates[1]
    })

df_geo = pd.DataFrame(rows)
df_geo
````
