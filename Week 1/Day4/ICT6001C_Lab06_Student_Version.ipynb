## Part 2: Loss calculation


We begin by computing the loss of the model.

Recall the regression model:

$$
\hat{y} = wx + b
$$

We measure prediction error using **Half Mean Squared Error (Half MSE)**:

$$
L(w,b) = \frac{1}{2n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

The factor 1/2 simplifies the derivative when computing gradients.

## Part 3 Backpropagation

To improve the model, we compute the gradients of the loss with respect to the parameters.

This step is called backpropagation because the prediction error is propagated backward to update the parameters.

---

### Task 3.1 Gradient with Respect to \( w \)

The gradient of the loss with respect to \( w \) is:

$$
\frac{\partial L}{\partial w}
=
-\frac{1}{n}\sum_{i=1}^{n} x_i (y_i - \hat{y}_i)
$$

Compute

$$
\frac{\partial L}{\partial w}
$$

using the data and predictions from Part 2.

---

### Task 3.2 Gradient with Respect to \( b \)

The gradient with respect to \( b \) is:

$$
\frac{\partial L}{\partial b}
=
-\frac{1}{n}\sum_{i=1}^{n} (y_i - \hat{y}_i)
$$

Compute

$$
\frac{\partial L}{\partial b}
$$

## Part 4 Gradient Descent

Once we know the gradients, we update the parameters using **gradient descent**.

Gradient descent moves parameters in the direction that reduces the loss.

The gradient points toward increasing loss, so we move in the opposite direction (hence we negate / minus).

The update rules are:

$$
w_{new} = w - \alpha \frac{\partial L}{\partial w}
$$

$$
b_{new} = b - \alpha \frac{\partial L}{\partial b}
$$

where

- $\alpha$ is the learning rate.

The learning rate controls the step size of the update.

For this exercise:

$$
\alpha = 0.1
$$
