## 1. Gini Impurity Formula

The Gini impurity of a dataset $(S)$ is defined as:

$$
\text{Gini}(S) = 1 - \sum_{k=1}^{K} p_k^2
$$

where $(p_k)$ is the proportion of samples in class $(k)$.


For a split of dataset \( S \) into subsets \( S_1, S_2, \dots, S_m \), the weighted Gini impurity is:

$$
\text{Gini}_{\text{split}} = \sum_{j=1}^{m} \frac{|S_j|}{|S|} \, \text{Gini}(S_j)
$$

The best split is the one that minimizes $\text{Gini}_{\text{split}}$.

## 3. Gini Tasks

1. Compute the Gini impurity at the root node.
2. At the current node, evaluate each candidate feature by:
   - Splitting the data into Yes/No groups
   - Computing the Gini impurity for each group
   - Computing the weighted Gini for the split
3. Select the feature with the lowest weighted Gini as the split.
4. Repeat Steps 2–3 for each resulting child node using the remaining features.
5. Stop splitting when the node is pure, no Gini reduction is achieved, or depth 2 is reached.

## 4. Entropy Tasks

1. Compute the entropy at the root node.
2. At the current node, evaluate each candidate feature by:
   - Splitting the data into Yes/No groups
   - Computing the entropy for each group
   - Computing the weighted entropy for the split
3. Select the feature with the lowest weighted entropy as the split.
4. Repeat Steps 2–3 for each resulting child node using the remaining features.
5. Stop splitting when the node is pure, no entropy reduction is achieved, or depth 2 is reached.
