# Lab 07: Unsupervised Learning (Clustering)

In this lab, you will study clustering from first principles and develop a deeper understanding of how structure can be discovered in unlabeled data. You will work through distance measures, clustering mechanics, and key concepts such as cluster assignment, centroid updates, and evaluation metrics.

The lab is totally pen and paper and you should not be using your laptop or phones. Please have only pen, paper and calculator on hand.

## Learning Objectives

By the end of this lab, you should be able to:

1. Compute and compare distance measures (L1, L2, and L∞) and explain their impact on similarity.
2. Perform K-means clustering manually, including cluster assignment and centroid updates.
3. Compute and interpret the K-means objective function (within-cluster sum of squares).
4. Evaluate clustering quality using the silhouette score and explain its meaning.
5. Explain how DBSCAN forms clusters and compare it with K-means in terms of assumptions and use cases.

### Task 2.5: Compute the silhouette score (Lecture 11 Slides 30-31)

For each point, compute:
$s(i)=\frac{b(i)-a(i)}{\max(a(i),b(i))}$

After computing $s(i)$ for each point, take the average of all $s(i)$ values to obtain the overall silhouette score.

Note: Silhouette score is computed using distances between data points, not distances to centroids.

| Point | d(μ₁ = (1.5,1)) | d(μ₂ = (4.5,3.5)) | Assigned Cluster |
